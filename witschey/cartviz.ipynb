```python
%matplotlib inline
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

### plot scatterplot

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

colors = {'best': 'blue', 'random': 'red'}
markers = {'best': 's', 'random': 'o'}
seen = {'best': False, 'random': False}

for d in data:
    kwargs = {'c': colors[splitter(d)],      # color
              'marker': markers[splitter(d)], # marker symbol
              's': 40,     # markersize
              'alpha': 0.7} # transparency
    if not seen[splitter(d)]:
        kwargs['label'] = splitter(d)
        seen[splitter(d)] = True
    ax.scatter(min_samples_split(d), max_leaf_nodes(d), median(d),
               **kwargs)

ax.legend()

ax.set_ylabel('min_samples_split')
ax.set_xlabel('max_leaf_nodes')
ax.set_zlabel('median')

plt.show()
```
